```python
# rename variable names with a function:
def rename_stock_columns(dataframe, prefix):
    # Define the new column names
    new_column_names = {
        'Open': prefix + '_open',
        'High': prefix + '_high',
        'Low': prefix + '_low',
        'Close': prefix + '_close',
        'Adj Close': prefix + '_adjclose',
        'Volume': prefix + '_vol'
    }
    # Rename the columns
    dataframe.rename(columns=new_column_names, inplace=True)

# Rename the columns for apple_stock_data and sp500_stock_data
rename_stock_columns(apple_stock_data, 'apple')
rename_stock_columns(sp500_stock_data, 'sp500')
```
